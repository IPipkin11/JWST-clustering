```python
import scipy
import scipy.stats
```

```python
radec = numpy.array([mastercat['ALPHA_J2000']/u.degree, mastercat['DELTA_J2000']/u.degree])
print(radec.shape)
```

```python
kde = scipy.stats.gaussian_kde(dataset=radec)
kde.covariance_factor = lambda : .25
kde._compute_covariance()
```

```python
values = numpy.array([mastercat['ALPHA_J2000'], mastercat['DELTA_J2000']])
kernel = scipy.stats.gaussian_kde(values)
weights = kernel(values)
weights = weights/weights.max()
```

```python
ra_min = 154
ra_max = 161
dec_min = -38
dec_max = -32.5
```

```python
ogRA, ogDEC = numpy.mgrid[ra_min:ra_max:100j, dec_min:dec_max:100j]
```

```python
og_RaDec = numpy.vstack([ogRA.ravel(), ogDEC.ravel()])
print(og_RaDec.shape)
smoothed = kde(og_RaDec)
smoothed2d = smoothed.reshape(ogRA.shape)
print(smoothed2d.shape)
```

```python
plt.figure(figsize=(10,10), dpi=100)
plt.imshow(numpy.rot90(smoothed2d),extent=[ra_min, ra_max, dec_min, dec_max], cmap ='gist_earth_r')
plt.scatter(mastercat['ALPHA_J2000'], mastercat['DELTA_J2000'], s=2, c = weights, alpha=0.6)

plt.scatter(157.5146, -35.3234, s=30, marker="o", c='red', zorder=0, label='Antlia Cluster Center')

            
plt.xlim((ra_max, ra_min))
plt.ylim((dec_min,dec_max))
plt.title("Kernel density estimation of UDG positions")
plt.xlabel('Right Ascension (J2000)')
plt.ylabel('Declination (J2000)')
plt.legend(fontsize=15)
#plt.tight_layout()
plt.colorbar(label='UDG candidate density', fraction=0.044)
plt.show()
```

```python
radius_bins = numpy.linspace(0, 2.5, 26)
print(radius_bins)
hist,bins = numpy.histogram(mastercat['distance2center'], bins=radius_bins)
hist_error = numpy.sqrt(hist)
```

```python
radius_bin_center = (radius_bins[1:] + radius_bins[:-1]) / 2.
print(radius_bin_center)

bin_area = 4 * numpy.pi * (radius_bins[1:]**2 - radius_bins[:-1]**2)
```

```python
galaxy_density = hist/bin_area
galaxy_density_error = hist_error / bin_area
```

```python
background_density = numpy.mean(galaxy_density[(radius_bin_center > 1.7) & (radius_bin_center < 2.0)] )
print(background_density)
```

```python
scale = 0.68 # mpc/degree
```

```python
bg_corrected_density = galaxy_density - background_density
density_error_with_bg = numpy.sqrt(galaxy_density_error**2 + 5**2)
plt.figure()
plt.scatter(radius_bin_center * scale, bg_corrected_density / scale**2)
plt.errorbar(x=radius_bin_center * scale, y=bg_corrected_density / scale**2, yerr=density_error_with_bg/scale**2)
plt.ylim((1,500))
plt.ylabel("Surface Density [Counts/Mpc**2]")
plt.xlabel("Projected Cluster-Center Radius [Mpc]")
plt.title('Radial Profile Plot of UDG candidates')
#plt.yscale('log')
plt.show()
```

```python
# cds['ra'] = 15. * (cds['RAh'] + cds['RAm']/60. + cds['RAs']/3600)
# cds['dec'] = cds['DEd'] + cds['DEm']/60. + cds['DEs']/3600.
df  = cds.to_pandas()
# df.info()
df['ra'] = 15. * (df['RAh'] + df['RAm']/60. + df['RAs']/3600)
df['dec'] = df['DEd'] + df['DEm']/60. + df['DEs']/3600.
# cds['ra'] = cds['RAh'] + cds['RAm'] + cds['RAs']
# cds['dec'] = cds['DEd'] + cds['DEm'] + cds['DEs']
# cds
print(numpy.array(df['ra']))
coma_sky = SkyCoord(numpy.array(df['ra'])*u.degree, numpy.array(df['dec'])*u.degree)
coma = SkyCoord(194.953054*u.degree, 27.980694*u.degree)
coma_r = coma.separation(coma_sky)
df['distance2center'] = coma_r / u.degree
```

```python
coma_bins = numpy.linspace(0,3.5,15)
coma_bin_center = 0.5 * (coma_bins[1:] + coma_bins[:-1])
print(coma_bins)
coma_bin_area = 4 * numpy.pi * (coma_bins[1:]**2 - coma_bins[:-1]**2)
coma_hist,bins = numpy.histogram(df['distance2center'], bins=coma_bins)
```

```python
coma_density = coma_hist / coma_bin_area
```

```python
coma_scale = 1.8
plt.figure()
plt.scatter(coma_bin_center*coma_scale, coma_density/coma_scale**2)
print(coma_density/coma_scale**2)
plt.show()
```
